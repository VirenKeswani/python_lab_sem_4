```python
#pandas progoram to add , sub , multiply , divide two pandas series
import pandas as pd
a = pd.Series([1, 2, 3, 4, 5])
b = pd.Series([1, 2, 3, 16, 25])
print('add\n',a.add(b),sep='')

print('sub\n',a.sub(b),sep='')

print('multiply\n',a.mul(b),sep='')

print('divide\n',a.div(b),sep='')

#compare elements of 2 pandas series
print('compare elements of 2 pandas series\n',a.eq(b),sep='')

#convert a dictionary to a pandas series
d = {'a': 6, 'b': 9, 'c': 7}
print('convert a dictionary to a pandas series\n',pd.Series(d),sep='')

#convert numpy array to pandas series
import numpy as np
a = np.array([5, 3, 2, 4, 5])
print('convert numpy array to pandas series\n',pd.Series(a),sep='')




```

```python
#wirte a program to read csv file in a dataframe and replace missing values with any value m drop the row if all values are missing or contain null values

df = pd.read_csv('SteamCharts.csv')
print('read csv file in a dataframe\n',df,sep='')
```

```python
df.replace('^\s+$','0000',regex=True,inplace=True)

print('replace missing values with any value\n',df,sep='')

```

```python
#remove with null
df.dropna(inplace=True)
print('remove with null\n',df,sep='')
```

```python
#Write a program to demonstrate merging of Frames: 
# i) on the basis of id
# ii) using how 

df1 = pd.DataFrame({'id': [1, 2, 3, 4, 5], 'name': ['a', 'b', 'c', 'd', 'e']})
df2 = pd.DataFrame({'id': [1, 2, 3, 4, 5], 'surname': ['f', 'g', 'h', 'i', 'j']})

#id to merge dataframes
print('id to merge dataframes\n',df1.merge(df2, on='id'),sep='')

#merge using how
print('merge using how\n',df1.merge(df2, how='outer'),sep='')
```
