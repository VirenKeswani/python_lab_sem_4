```python
_input = [int(item) for item in input().split()]
# put even odd numbers in two lists

# make menu
while True:
    print("""
    1. Seperate Even Odd
    2. Merge and Sort
    3. Update First element
    4. Print Middle Element
    5. Exit""")


    choice = int(input("Enter your choice: "))
    if choice == 1:
        even, odd = [], []
        for i in _input:
            if i % 2 == 0:
                even.append(i)
            else:
                odd.append(i)
        print("Even:", even)
        print("Odd:", odd)
    
    elif choice == 2:
        new_list = [int(item) for item in input().split()]
        _input += new_list
        print("Merged: ", _input)
        print("Sorted: ", sorted(_input))

    elif choice == 3:
        _input[0] = int(input("Enter new first element: "))
        print("Updated list: ", _input)
    
    elif choice == 4:
        print("Middle element: ", _input[int(len(_input)//2)])

    elif choice == 5:
        break

    else:
        print("Invalid choice")
```

```python
student_data = []

def new_student():
    temp_name = input("Enter name: ")
    temp_score_m = float(input("Enter score for math: "))
    temp_score_s = float(input("Enter score for science: "))
    temp_score_e = float(input("Enter score for english: "))
    temp_id = int(input("Enter id: "))
    
    student_data.append((temp_name, temp_score_m, temp_score_s, temp_score_e, temp_id))


def get_student(name):
    for student in student_data:
        if student[0] == name:
            return student
    
    return None

# make menu
while True:
    print("1. New student")
    print("2. Get student")
    print("3. Exit")
    _input = int(input())

    if _input == 1:
        new_student()
    elif _input == 2:
        arg = {}
        print("Enter name: ")
        arg = input()
        student = get_student(arg)
        if student is None:
            print("Student not found")
        else:
            print(student)
    elif _input == 3:
        break
    else:
        print("Invalid input")
```

```python
# menu driven for sets

A = set(input("Enter A: ").split())
B = set(input("Enter B: ").split())

while True:
    print("""
1. A union B
2. A intersection B
3. A difference B
4. A symmetric difference B
5. Exit""")

    choice = int(input("Enter your choice: "))
    if choice == 1:
        print(A | B)
    elif choice == 2:
        print(A & B)
    elif choice == 3:
        print(A - B)
    elif choice == 4:
        print(A ^ B)
    elif choice == 5:
        break
```

```python
# all about dicts
len_dict = int(input("Enter length of dict: "))
my_dict = {}
for i in range(len_dict):
    key = input("Enter key: ")
    value = input("Enter value: ")
    my_dict[key] = value

# sort dict by key
my_dict = sorted(my_dict.items())

print("Sorted dict: ", my_dict)

# input new dict and update old one
new_dict = {}

for i in range(len_dict):
    key = input("Enter key: ")
    value = input("Enter value: ")
    new_dict[key] = value

my_dict.update(new_dict)

print("Updated dict: ", my_dict)
```
